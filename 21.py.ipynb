```python
def process_action(monkey,known_monkeys):
    if monkey not in known_monkeys:
        s = cdf.action[monkey]
        assert(len(s.split())== 3)
        a,operator,b = s.split()
        if a in known_monkeys:
            if b in known_monkeys:
                if operator == '+':
                    known_monkeys[monkey] = known_monkeys[a] + known_monkeys[b]
                    #print('Success!')
                elif operator == '-':
                    known_monkeys[monkey] = known_monkeys[a] - known_monkeys[b]
                    #print('Success!')        
                elif operator == '*':
                    known_monkeys[monkey] = known_monkeys[a] * known_monkeys[b]
                    #print('Success!')                    
                elif operator == '/':
                    known_monkeys[monkey] = known_monkeys[a] / known_monkeys[b]
                    #print('Success!')                    
                else:
                    print("WARNING: Unknown operator {}".format(operator))
            #else:
                #print("{} not known yet".format(b))
        #else:
            #print("{} not known yet".format(a))
    #else:
    #    print("Monkey {} already shouted {} ".format(monkey,known_monkeys[monkey]))
                
                    
    
    
    
```

```python
process_action('qppn',pain)
```

```python
for i in range(100):
    for m in cdf.index:
        process_action(m,pain)
```
